Smoothing Images

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

2D Convolution (Image Filtering)

```python
img = cv2.imread('ml.png')

kernel = np.ones((5,5),np.float32)/25
dst = cv2.filter2D(img,-1,kernel)

plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(dst),plt.title('Averaging')
plt.xticks([]), plt.yticks([])
plt.show()
```

Gaussian Filtering

```python
img = cv2.imread('ml.png')
blur5 = cv2.blur(img,(15,15),0)
GaussianBlur5 = cv2.GaussianBlur(img,(15,15),0)

plt.subplot(121),plt.imshow(blur5),plt.title('blur5')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(GaussianBlur5),plt.title('Gaussianblur5')
plt.xticks([]), plt.yticks([])
plt.show()
```

Median Filtering -  5  

```python
img = cv2.imread('ml.png')
blur5 = cv2.blur(img,(15,15),0)
median = cv2.medianBlur(img,15)

plt.subplot(121),plt.imshow(blur5),plt.title('blur5')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(median),plt.title('median')
plt.xticks([]), plt.yticks([])
plt.show()
```

Bilateral Filtering

```python
img = cv2.imread('ml.png')
blur5 = cv2.blur(img,(15,15))
bilateral = cv2.bilateralFilter(img,9,75,75)

plt.subplot(121),plt.imshow(blur5),plt.title('blur5')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(bilateral),plt.title('bilateral')
plt.xticks([]), plt.yticks([])
plt.show()
```
