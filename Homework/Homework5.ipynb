```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img1 = cv2.imread('car.png',0)
img2 = cv2.imread('123.png',0)

orb = cv2.ORB_create()
kp1, des1 = orb.detectAndCompute(img1, None)
kp2, des2 = orb.detectAndCompute(img2, None)

#sift = cv2.SIFT_create()
#kp1, des1 = sift.detectAndCompute(img1,None)
#kp2, des2 = sift.detectAndCompute(img2,None)

bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)
matches = bf.match(des1, des2)
matches = sorted(matches, key = lambda x:x.distance)

#bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)
#matcher = cv2.BFMatcher(cv2.NORM_L1, crossCheck=True)
#matches = matcher.match(des1,des2)

matching_result = cv2.drawMatches(img1, kp1, img2, kp2, matches[:100], None, flags=2)
plt.imshow(matching_result), plt.show()
cv2.imwrite('Homework5.png',matching_result)
#matches = sorted(matches, key = lambda x:x.distance)
```
